```python
class Barco:
    def __init__(self, nombre, tamanno):
        self.nombre = nombre
        self.tamanno = tamanno
    
    def slora(self):
        "Crea un array de numpy con el tamaño del barco en letras 'b'"
        return np.full(self.tamanno, "b", dtype=str)
    
    def fletar(self,orientacion, altitud, longitud, tablero_iniciado):
        if orientacion == 0:
            tablero_iniciado[2, altitud:altitud +self.tamanno, longitud] = self.slora()
        else:
            tablero_iniciado[2, altitud, longitud:longitud +self.tamanno] = self.slora()

TIPOS_BARCOS = {
    # 1: (Barco("Txalupa", 1), 4),
    # 2: (Barco("Bergantin", 2), 3),
    # 3: (Barco("Carabela", 3), 2),
    4: (Barco("Galeón", 4), 1)
}
```

```python
def apuntar(tablero_iniciado):
    '''
    Entrada del juego, con validación rudimentaria de la entrada del usuario.
    '''
    i = int(input("i: "))
    j = int(input("j: "))
    
    while True:
        if i < 1 or i > 10 or j < 1 or j > 10:
            print("¡Fuera de rango!")
            i = int(input("i: "))
            j = int(input("j: "))
        else:
            disparo(i, j, tablero_iniciado)
            break


def disparo(i, j, tablero_iniciado):
    '''
    Recive las coordenadas de la función entrada() para comprobar si en el sitio
    había o no un barco y volver a efectuar otro disparo o no.
    '''
    if tablero_iniciado[0, i -1, j -1] == "b" or tablero_iniciado[1, i -1, j -1] == "d":
        print("¡Tocado!")
        tablero_iniciado[1, i -1, j -1] = "x"
        apuntar(tablero_iniciado)
    else:
        print("¡Agua!")
        tablero_iniciado[1, i -1, j -1] = "O"
    
# apuntar(np.full(((3, 10, 10)), "_", dtype=str))
```

```python
def entrada(tablero_iniciado):
    while any(cantidad > 0 for _, (_, cantidad) in TIPOS_BARCOS.items()): # _ Representa un valor a ignorar, _ Clave y _ barco
        
        print("Barcos en el dique")
        
        for clave, (barco, cantidad) in TIPOS_BARCOS.items():
            if cantidad > 0:
                print(f"{clave}: {barco.nombre} tamaño: {barco.tamanno} - cantidad: {cantidad}")
        
        barco_seleccion = int(input("Elige un barco (Galeon(4), Bergantin(3), Carabela(2), Txalupa(1)): (0 para salir) "))
        if barco_seleccion == 0:
            break

        if barco_seleccion not in TIPOS_BARCOS or TIPOS_BARCOS[barco_seleccion][1] <= 0:
            print("Selección no válida o barco agotado.")
            continue
        
        barco, cantidad = TIPOS_BARCOS[barco_seleccion]
        if cantidad <= 0:
            print(f"No quedan {barco.nombre}s disponibles.")
            return
                    
        orientacion = int(input("Vertical(0) / Horizontal(1): "))
        altitud = int(input("Altitud del barco: ")) -1
        longitud = int(input("Longitud del barco: ")) -1
        
        barco.fletar(orientacion, altitud, longitud, tablero_iniciado)
        TIPOS_BARCOS[barco_seleccion] = (barco, cantidad - 1)
        apuntar(tablero_iniciado)
        
# entrada(np.full(((3, 10, 10)), "_", dtype=str))
```

```python
def iniciar():
    # clear_output(wait=True)
    z = int(input("Introduce el tamaño del tablero: "))
    tablero = np.full(((3, z, z)), "_", dtype=str)
    entrada(tablero)
    apuntar(tablero)
iniciar()
```

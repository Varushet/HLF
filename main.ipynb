```python
def tablero(z):
    tablero = np.full(((3, z, z)), "_", dtype=str)
    return tablero

# tablero(10)
```

```python
def flotar(tablero_iniciado):   
    
    galeon = np.array(["b", "b", "b", "b"])
    bergantin = np.array(["b", "b", "b"])
    carabela = np.array(["b", "b"])
    txalupa = np.array(["b"])
    
    orientacion = int(input("Vertical(0) / Horizontal(1): "))
    barco = int(input("Elige un barco (Galeon(4), Bergantin(3), Carabela(2), Txalupa(1)): "))
    dique = 0
    
    if barco == 3:
        dique = 2
    elif barco == 2:
        dique = 3
    elif barco == 1:
        dique = 4
        
    for _ in range(dique):
        match barco:
            case 4:
                tamaño = galeon
            case 3:
                tamaño = bergantin
            case 2:
                tamaño = carabela
            case 1:
                tamaño = txalupa
            case 0:
                return
            
        altitud = int(input("Altitud del barco: ")) -1
        longitud = int(input("Longitud del barco: ")) -1
        
        if orientacion == 0:
            clear_output(wait=True)
            tablero_iniciado[0, altitud:altitud +barco, longitud] = tamaño
        else:
            clear_output(wait=True)
            tablero_iniciado[0, altitud, longitud:longitud +barco] = tamaño
    
    print(tablero_iniciado[0])
    
# flotar(tablero(10))
```

```python
def entrada():
    '''
    Entrada del juego, con validación rudimentaria de la entrada del usuario.
    '''
    i = int(input("i: "))
    j = int(input("j: "))
    
    while True:
        if i < 1 or i > 10 or j < 1 or j > 10:
            print("¡Fuera de rango!")
            i = int(input("i: "))
            j = int(input("j: "))
        else:
            disparo(i, j)
            break


def disparo(i, j):
    '''
    Recive las coordenadas de la función entrada() para comprobar si en el sitio
    había o no un barco y volver a efectuar otro disparo o no.
    '''
    if tablero[i -1, j -1] == "b":        
        print(tablero)
        print("¡Tocado!")
        tablero[i -1, j -1] = "x"
        entrada()
        return True
    else:
        print(tablero)
        print("¡Agua!")
        tablero[i -1, j -1] = "O"
        return False
    
entrada()
```

```python
def iniciar():
    z = int(input("Introduce el tamaño del tablero: "))
    tablero_iniciado = tablero(z)
    flotar(tablero_iniciado)
    

iniciar()
```
